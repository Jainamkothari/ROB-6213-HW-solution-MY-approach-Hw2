$$\begin{align}
    x_{n+1} = A
    x_n + B u_n
\end{align}$$

$u_n$ contains the forces produced by the rotors (our control inputs).
The module ```quadrotor.py``` defines the matrices $A$ and $B$ and code to generate a simulation of the drone.

We want to generate a control input will move the drone towards the point $(3,3)$ starting from $(0,0)$. In order to
do so, we define a cost that penalizes the distance to the goal at each time step while minimizing velocities and the amount of trust needed,
i.e. we want to solve the following optimal control problem
$$\begin{align}
&\min_{x_0, u_0, x_1, u_1, \cdots} \frac{1}{2}\sum_{n=0}^{N} (x_n - x_{desired})^T Q (x_n - x_{desired}) + u_n^T R u_n\\
\textrm{subject to}\ \ & x_{n+1} = A x_{n} + B u_n\\
& x_0 = [0,0,0,0,0,0]^T
\end{align}$$
where $x_{desired} = [3,0,3,0,0,0]^T$

1. Write down the KKT conditions for the problem
2. Write code to solve the problem for N=500 time steps (you will need to find diagonal matrices $Q>0$ and $R>0$ to create a nice movement) by solving the KKT system of linear equations. Do NOT inverse the KKT matrix, instead use the [NumPy solve function](https://numpy.org/doc/stable/reference/generated/numpy.linalg.solve.html) which is numerically more stable and efficient.
4. Show plots of all the states of the robot as a function of time

````python
## LAST HOMEWORK



import numpy as np

# Define constants
N = 500 # Number of time steps
x_dim = 6  # Dimension of state vector
u_dim = 2  # Dimension of control vector

# Define the state cost matrix Q and the control effort cost matrix 
Q = np.diag([30, 20, 30, 20, 1, 0])  # Penalize position, velocity
R = np.diag([0.5, 0.5])  #Pena

# Desired state target
x_desired = np.array([3, 0, 3, 0, 0, 0])

# Construct the vector g^T as we can see in the equations above and taught in class
g = np.tile(np.hstack((((-x_desired) @ Q), [0]*u_dim)), N)
g = g.reshape(g.shape[0], 1)  
g = g.astype(np.float64)

# Construct the vector p, which contains the initial state and zeros for the constraints
p = np.zeros(((x_dim*N), 1))
p = p.astype(np.float64)


# Initialize G matrix
G = np.zeros(((x_dim + u_dim) * N, (x_dim + u_dim) * N))

# Fill G matrix with blocks of Q and R
for i in range(N):
    start_row = i * (x_dim + u_dim)
    end_row = start_row + (x_dim + u_dim)
    
    G[start_row:start_row + x_dim, start_row:start_row + x_dim] = Q  # Insert Q block
    G[start_row + x_dim:start_row + (x_dim + u_dim), start_row + x_dim:start_row + (x_dim + u_dim)] = R  # Insert R block

# Convert G to float64
G = G.astype(np.float64)

# Initialize C with zeros 
C = np.zeros(((x_dim + (N - 1) * x_dim), (x_dim + u_dim) * N))

# Fill in the initial condition for row 0
C[:x_dim, :x_dim] = np.eye(x_dim)  # Initial condition

# Iterate through each time step and fill A, B, and -I
for i in range(1, N):  # Start from 1 because the first row is the initial condition
    row_start = i * x_dim  # Each block of dynamics occupies x_dim rows

    # Assign A
    C[row_start:row_start + x_dim, (i - 1) * (x_dim + u_dim):(i - 1) * (x_dim + u_dim) + x_dim] = quadrotor.A
    
    # Assign B
    C[row_start:row_start + x_dim, (i - 1) * (x_dim + u_dim) + x_dim:(i - 1) * (x_dim + u_dim) + (x_dim + u_dim)] = quadrotor.B
    
    # Assign -I
    C[row_start:row_start + x_dim, (i - 1) * (x_dim + u_dim) + (x_dim + u_dim):(i - 1) * (x_dim + u_dim) + (x_dim + u_dim) + x_dim] = -np.eye(x_dim)

# Convert C to float64
C = C.astype(np.float64)

#Construct the left-hand side (LHS) matrix
rows_G, cols_G = G.shape
rows_C, cols_C = C.shape
rows_CT, cols_CT = C.T.shape  # Get dimensions for C^T

# Initialize LHS
KKT = np.zeros((rows_G + rows_C, cols_G + cols_CT))  # Create an empty matrix for LHS

# Fill in G into LHS (top-left)
KKT[:rows_G, :cols_G] = G

# Fill in C^T into LHS (top-right)
KKT[:rows_CT, cols_G:cols_G + cols_CT] = C.T

# Fill in C into LHS (bottom-left)
KKT[rows_G:rows_G + rows_C, :cols_C] = C

# The bottom-right portion of LHS will be zeros already

# Convert LHS to float64 if needed
KKT = KKT.astype(np.float64)

# Calculate size for Final_desired
total_rows = g.shape[0] + p.shape[0]
Final_desired = np.zeros((total_rows, 1), dtype=np.float64)

# Fill in values from -g
Final_desired[:g.shape[0], 0] = -g[:, 0]  # Negate g and fill in

# Fill in values from p
Final_desired[g.shape[0]:, 0] = p[:, 0]  # Fill in p


# Solve the linear system (LHS * [y, lambda] = RHS) to find the optimal y and lambda
res = np.linalg.solve(KKT, Final_desired)

# Extract the optimal states (x) and controls (u) from the solution vector
# Reshape the first 4000 elements (corresponding to x and u values) into blocks of size 8
# Extract and transpose x values to shape (6, 500)
x1 = res[:(x_dim + u_dim) * N].reshape(-1, 8)[:, :6].T
# Extract and transpose u values to shape (2, 500)
u1 = res[:(x_dim + u_dim) * N].reshape(-1, 8)[:, 6:].T

# Animate the robot (quadrotor) using the computed states and controls
quadrotor.animate_robot(x1, u1, goal=[3, 3])

# Time vector for plotting
time_steps = np.arange(N)

# Plot all states of the robot as a function of time
plt.figure(figsize=(12, 8))
plt.subplot(2, 1, 1)
plt.plot(time_steps, x1[0], label='Position X')
plt.plot(time_steps, x1[1], label='Velocity X')
plt.plot(time_steps, x1[2], label='Position Y')
plt.plot(time_steps, x1[3], label='Velocity Y')
plt.plot(time_steps, x1[4], label='Position Z')
plt.plot(time_steps, x1[5], label='Velocity Z')
plt.title('Robot States Over Time')
plt.xlabel('Time Steps')
plt.ylabel('State Values')
plt.legend()
plt.grid()

# Plot the optimal control as a function of time
plt.subplot(2, 1, 2)
plt.plot(time_steps, u1[0], label='Control U1', color='orange')
plt.plot(time_steps, u1[1], label='Control U2', color='green')
plt.title('Optimal Control Inputs Over Time')
plt.xlabel('Time Steps')
plt.ylabel('Control Values')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
````
